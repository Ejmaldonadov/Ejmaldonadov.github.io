```python
import pandas as pd
datos = {
    'Usuario': ['Josué', 'Josué', 'Pedro', 'Pedro', 'Maria', 'Maria'],
    'Contenido': ['Marvel', 'DC', 'DC', 'StarWars', 'Marvel', 'StarWars'],
    'Calificacion': [5, 4, 5, 2, 3, 4]
}

# Crear DataFrame
df = pd.DataFrame(datos)

# Mostrar los datos
print(df)

print("-----------------------------")
# Crear la matriz de usuario-contenido
matriz = df.pivot_table(index='Usuario', columns='Contenido', values='Calificacion')

# Mostrar la matriz
print(matriz)

print("------------------------------")

from sklearn.metrics.pairwise import cosine_similarity

# Rellenar NaN con 0 para el cálculo de similitud
matriz_usuarios = matriz.fillna(0)

# Calcular la similitud del coseno entre usuarios
similitud = cosine_similarity(matriz_usuarios)

# Crear DataFrame con los resultados
similitud_df = pd.DataFrame(similitud, index=matriz_usuarios.index, columns=matriz_usuarios.index)

# Mostrar la matriz de similitud
print(similitud_df)

print("-------------------------------")

import numpy as np

# Seleccionar usuario objetivo
usuario_objetivo = 'Josué'

# Obtener series: calificaciones del usuario objetivo y sus similitudes
calificaciones_usuario = matriz.loc[usuario_objetivo]
similitudes = similitud_df[usuario_objetivo]

# Crear lista para almacenar predicciones
predicciones = {}

# Para cada contenido que no ha calificado Josué
for contenido in matriz.columns:
    if pd.isna(calificaciones_usuario[contenido]):
        # Obtener calificaciones de otros usuarios para ese contenido
        calificaciones_contenido = matriz[contenido]

        # Calcular predicción: promedio ponderado por similitud
        numerador = 0
        denominador = 0

        for otro_usuario in matriz.index:
            if otro_usuario != usuario_objetivo and not pd.isna(calificaciones_contenido[otro_usuario]):
                sim = similitudes[otro_usuario]
                calif = calificaciones_contenido[otro_usuario]
                numerador += sim * calif
                denominador += sim

        if denominador != 0:
            prediccion = numerador / denominador
            predicciones[contenido] = prediccion

recomendaciones = sorted(predicciones.items(), key=lambda x: x[1], reverse=True)
print("Recomendaciones para", usuario_objetivo)
for contenido, score in recomendaciones:
    print(f"{contenido}: {score:.2f}")

usuario_objetivo = 'Pedro'
calificaciones_usuario = matriz.loc[usuario_objetivo]
similitudes = similitud_df[usuario_objetivo]
predicciones = {}

for contenido in matriz.columns:
    if pd.isna(calificaciones_usuario[contenido]):
        calificaciones_contenido = matriz[contenido]

        numerador = 0
        denominador = 0

        for otro_usuario in matriz.index:
            if otro_usuario != usuario_objetivo and not pd.isna(calificaciones_contenido[otro_usuario]):
                sim = similitudes[otro_usuario]
                calif = calificaciones_contenido[otro_usuario]
                numerador += sim * calif
                denominador += sim

        if denominador != 0:
            prediccion = numerador / denominador
            predicciones[contenido] = prediccion


recomendaciones = sorted(predicciones.items(), key=lambda x: x[1], reverse=True)
print("Recomendaciones para", usuario_objetivo)
for contenido, score in recomendaciones:
    print(f"{contenido}: {score:.2f}")

usuario_objetivo = 'Maria'
calificaciones_usuario = matriz.loc[usuario_objetivo]
similitudes = similitud_df[usuario_objetivo]

predicciones = {}
for contenido in matriz.columns:
    if pd.isna(calificaciones_usuario[contenido]):

        calificaciones_contenido = matriz[contenido]

        numerador = 0
        denominador = 0

        for otro_usuario in matriz.index:
            if otro_usuario != usuario_objetivo and not pd.isna(calificaciones_contenido[otro_usuario]):
                sim = similitudes[otro_usuario]
                calif = calificaciones_contenido[otro_usuario]
                numerador += sim * calif
                denominador += sim

        if denominador != 0:
            prediccion = numerador / denominador
            predicciones[contenido] = prediccion

recomendaciones = sorted(predicciones.items(), key=lambda x: x[1], reverse=True)
print("Recomendaciones para", usuario_objetivo)
for contenido, score in recomendaciones:
    print(f"{contenido}: {score:.2f}")
```
